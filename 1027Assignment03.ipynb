Basic Matrix Operations

```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns


# Create two 3Ã—3 matrices
M1 = np.random.randint(0, 10, (3, 3))
M2 = np.random.randint(0, 10, (3, 3))

# Addition
sum_mat = M1 + M2

# Subtraction
diff_mat = M1 - M2

# Multiplication
prod_mat = np.dot(M1, M2)

print("Matrix M1:\n", M1)
print("\nMatrix M2:\n", M2)

print("\nAddition:\n", sum_mat)
print("\nSubtraction:\n", diff_mat)
print("\nMultiplication:\n", prod_mat)


# Determinant of M2
det_val = np.linalg.det(M2)

# Inverse
if det_val != 0:
    inv_mat = np.linalg.inv(M2)
else:
    inv_mat = "Matrix is Singular, cannot compute"

# Rank
rank_val = np.linalg.matrix_rank(M2)

print("\nDeterminant of Matrix M2:", det_val)
print("\nInverse Matrix:\n", inv_mat)
print("\nRank of Matrix M2:", rank_val)
```

Matrix Plotting and
Visualization

```python
# Create two vectors with 15 random floats
vec_a = np.random.rand(15)
vec_b = np.random.rand(15)

plt.figure()
plt.plot(vec_a, marker='o', label='Vector A')
plt.plot(vec_b, marker='s', label='Vector B')

plt.xlabel("Index")
plt.ylabel("Value")
plt.title("Comparison of Two Random Vectors")
plt.legend()

plt.show()
```

```python
# Create 4x4 matrix
heat_matrix = np.random.rand(4, 4)

# Heatmap
sns.heatmap(
    heat_matrix,
    annot=True,
    cmap='copper',
    linewidths=1,
    linecolor='white'
)

plt.title('Heatmap of 4x4 Matrix')
plt.xlabel("Column Index")
plt.ylabel("Row Index")
plt.show()
```

```python
# Create two 4x4 matrices
MatX = np.random.rand(4, 4)
MatY = np.random.rand(4, 4)

# Operations
sum_result = MatX + MatY
diff_result = MatX - MatY
prod_result = np.dot(MatX, MatY)

# Plotting
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Addition
axes[0].bar(range(1, 17), sum_result.flatten(), color='blue')
axes[0].set_title('Matrix Addition')
axes[0].set_xlabel('Index')
axes[0].set_ylabel('Value')

# Subtraction
axes[1].bar(range(1, 17), diff_result.flatten(), color='green')
axes[1].set_title('Matrix Subtraction')
axes[1].set_xlabel('Index')
axes[1].set_ylabel('Value')

# Multiplication
axes[2].bar(range(1, 17), prod_result.flatten(), color='purple')
axes[2].set_title('Matrix Multiplication')
axes[2].set_xlabel('Index')
axes[2].set_ylabel('Value')

plt.tight_layout()
plt.show()
```
